# FP GROWTH

```python
## FP-Growth stands for frequent pattern growth. It is used to find frequent itemsets from transactional data without generating candidate itemsets repeatedly. Introduced by Jiawei Han in 2000, it was designed to overcome the performance bottlenecks of the Apriori algorithm, particularly the expensive process of candidate generation.



## how it works:
scan the transaction dataset and calculate support of each item.
remove infrequent items based on minimum support.
sort the remaining items by frequency.
build an fp-tree by inserting sorted transactions.
mine the fp-tree to extract frequent itemsets.
generate association rules from the frequent itemsets if needed.



important concepts:
support → how frequently an item or itemset appears in the dataset.
confidence → how often the rule is correct when the condition is true.
lift → strength of association between items compared to random occurrence.



FP-Growth stands for frequent pattern growth. It is used to find frequent itemsets from transactional data without generating candidate itemsets repeatedly. Introduced by Jiawei Han in 2000, it was designed to overcome the performance bottlenecks of the Apriori algorithm, particularly the expensive process of candidate generation.



how it works:
scan the transaction dataset and calculate support of each item.
remove infrequent items based on minimum support.
sort the remaining items by frequency.
build an fp-tree by inserting sorted transactions.
mine the fp-tree to extract frequent itemsets.
generate association rules from the frequent itemsets if needed.

    
    important concepts:
support → how frequently an item or itemset appears in the dataset.
confidence → how often the rule is correct when the condition is true.
lift → strength of association between items compared to random occurrence.

    
    important parameters:
min_support: minimum support required for an itemset to be frequent.
use_colnames=true: shows actual item names instead of column indexes.
association_rules(): used to generate rules from frequent itemsets.
metric='confidence': common metric for rule generation.
min_threshold: minimum value for the chosen metric.




how to choose important parameters:
if min_support is given in the question, use it directly.
if not given, choose a simple value like 0.3 or 0.4 depending on dataset size.
if no rules are generated, reduce support or confidence slightly.
if too many rules are generated, increase support or confidence.




real-world applications:
market basket analysis
product recommendation
customer buying pattern analysis
website click pattern analysis
medical and bioinformatics pattern discovery


advantages:
faster than apriori for large datasets
does not generate too many candidate itemsets
efficient for dense transactional data
gives meaningful frequent itemsets and rules


limitations:
fp-tree construction can become complex to explain manually
performance may reduce for very sparse or highly varied data
result quality depends on proper support/confidence selection
association does not always mean causation




CODE 


1] import the packages
import pandas as pd
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

import warnings
warnings.filterwarnings('ignore')


2] generate / obtain the dataset
# sample transaction data. in exam, replace these with the given transactions

transactions = [
    ['milk', 'bread', 'butter'],
    ['bread', 'butter', 'eggs'],
    ['milk', 'bread'],
    ['milk', 'eggs'],
    ['bread', 'butter'],
    ['milk', 'bread', 'eggs'],
    ['milk', 'butter'],
    ['bread', 'eggs']
]

# convert transaction list to binary basket format
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns = te.columns_)

print("encoded transaction data:\n")
df



3] check single item support
# this helps us understand which items appear more often
item_support = df.mean().sort_values(ascending=False)

print("\nsingle item support:")
print(item_support)







plt.figure(figsize=(6,4))

item_support.plot(kind='bar')

plt.title('single item support')
plt.ylabel('support')
plt.xlabel('items')
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()








# cell: optional fp-tree style visualization (graphviz)

# this is a simple conceptual visualization
# not a full fp-tree implementation (exam-safe shortcut)

from graphviz import Digraph

dot = Digraph()

# add root node
dot.node('root', 'root')

# create simple tree using transactions
for i, transaction in enumerate(transactions):
    parent = 'root'
    
    # sort items by global frequency (important step in fp-growth)
    sorted_items = sorted(transaction, key=lambda x: item_support[x], reverse=True)
    
    for item in sorted_items:
        node_id = parent + '_' + item
        
        dot.node(node_id, item)
        dot.edge(parent, node_id)
        
        parent = node_id

# display graph
dot













4] apply fp-growth
# if min_support is given in exam, use that directly
# otherwise, use a simple value like 0.3 or 0.4
frequent_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)

# sort for clean output
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)

print("frequent itemsets:\n")
print(frequent_itemsets)








5] generate association rules
# if confidence threshold is given in exam, use that directly
# otherwise, use 0.6 as a simple exam-safe value
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.6)

# keep only useful columns
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules = rules.sort_values(by = ['lift', 'confidence'], ascending = False)

print("association rules:\n")
print(rules)




6] simple interpretation-friendly output
print("\nnumber of frequent itemsets:", len(frequent_itemsets))
print("number of association rules:", len(rules))

# optional: show strongest rule if rules exist
if len(rules) > 0:
    best_rule = rules.iloc[0]
    print("\nbest rule based on lift/confidence:\n")
    print("antecedent:", best_rule['antecedents'])
    print("consequent:", best_rule['consequents'])
    print("support:", round(best_rule['support'], 3))
    print("confidence:", round(best_rule['confidence'], 3))
    print("lift:", round(best_rule['lift'], 3))
else:
    print("\nno rules found. reduce min_support or confidence threshold.")





7] final interpretation
Frequent itemsets show combinations of items that occur together often based on the chosen minimum support.
Items with higher support are more common in the dataset.
Association rules represent relationships between items in "if-then" form.
Higher confidence means the rule is more reliable, and lift > 1 indicates a strong positive association.
Overall, FP-Growth efficiently finds meaningful patterns without generating unnecessary candidate itemsets.
```

# PCA

```python
PCA is a dimensionality reduction technique used to transform a dataset with many features into a smaller set of new variables called principal components, while preserving as much variance as possible.

how it works:
standardize the dataset so all features are on the same scale.
find directions in the data where the variance is maximum.
create new axes called principal components.
rank the components based on explained variance.
keep the top components and reduce the dimensionality of the dataset.

    
    
    
important parameters / methods:
n_components: number of principal components to keep.
explained_variance_ratio_: shows the proportion of variance explained by each component.
cumulative explained variance: total variance retained by selected components.
standardscaler: used before pca when features are on different scales.

    
    
    
    
    how to choose important parameters:
if the number of components is given in the question, use it directly.
if not given, choose the number of components based on explained variance.
usually, enough components are selected to preserve around 90% to 95% of the total variance.
for visualization, 2 principal components are commonly used.

    
    
    
    
    real-world applications:
image compression
feature reduction before machine learning
noise reduction
pattern recognition
visualization of high-dimensional datasets
bioinformatics and finance data analysis

        
        
        
        advantages:
reduces the number of features
removes redundancy in data
helps in visualization
can improve efficiency of further analysis
keeps maximum variance in fewer dimensions

    
    
    
    limitations:
principal components are not always easy to interpret
some information is lost after reduction
pca works best for linear relationships
results can be affected if data is not standardized



1] import the packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')



2] generate / obtain the dataset

# iris is a standard dataset used for pca practical
iris = load_iris()

x = iris.data
y = iris.target

feature_names = iris.feature_names
target_names = iris.target_names

print("original data shape:", x.shape)





3] visualize raw data
plt.figure(figsize=(6, 4))

plt.scatter(x[:, 0], x[:, 1])

plt.title('raw data (before pca)')
plt.xlabel('feature 1')
plt.ylabel('feature 2')

plt.grid(alpha=0.3)
plt.show()



4] standardize the data
# pca is affected by scale, so standardization is important
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

df = pd.DataFrame(x_scaled, columns = feature_names)
df.head()





5] applying PCA
# reduce data to 2 principal components for easy visualization
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x_scaled)

print("PCA transformed data shape:", x_pca.shape)



pca_df = pd.DataFrame(x_pca, columns=['PC1', 'PC2'])
pca_df['species'] = y

print("first 5 rows of PCA data:\n")
pca_df.head()



6] explained variance
# this shows how much information each principal component keeps

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

print("explained variance ratio:", explained_variance)

print("\ncumulative explained variance:", cumulative_variance)



7] original data vs pca transformed data
# this compares original feature space with reduced pca space
# colors represent different classes for better visualization
plt.figure(figsize=(12, 5))

colors = ['red', 'green', 'blue']

datasets = [
    (x, 'original data', 'feature 1', 'feature 2'),
    (x_pca, 'data after pca', 'principal component 1', 'principal component 2')
]

for idx, (data, title, xlabel, ylabel) in enumerate(datasets):
    plt.subplot(1, 2, idx + 1)
    
    for i in range(len(target_names)):
        plt.scatter( data[y == i, 0], data[y == i, 1], color=colors[i],label=target_names[i])
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()



8] scree plot
# scree plot shows how much variance each principal component explains
# if the first few components explain most of the variance, they are enough
# after a certain point, the gain becomes small, so extra components may not be useful

plt.figure(figsize=(6, 4))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')

plt.title('scree plot')
plt.xlabel('principal components')
plt.ylabel('explained variance ratio')
plt.grid(True)
plt.show()




9] simple conclusion output
print("\nnumber of original features:", x.shape[1])
print("number of reduced features:", x_pca.shape[1])

print("\ninterpretation:\n")
print("PC1 explains:", round(explained_variance[0] * 100, 2), "% variance")
print("PC2 explains:", round(explained_variance[1] * 100, 2), "% variance")

print("\nfirst two principal components together explain:", round(cumulative_variance[-1] * 100, 2), "% variance")




10] optional: k-means on pca data with elbow method
# elbow method is used to choose k in k-means, not components in pca
# here we apply k-means on the reduced pca data

from sklearn.cluster import KMeans

inertia = []

for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(x_pca)
    inertia.append(model.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), inertia, marker='o')

plt.title('elbow method on pca data')
plt.xlabel('number of clusters')
plt.ylabel('inertia')

plt.grid(alpha=0.3)
plt.show()




11] optional: k-means clustering on pca data
# this is only used if the question asks to combine pca with clustering
# pca reduces dimensions first, then k-means is applied on the reduced data

kmeans = KMeans(n_clusters = 3, random_state = 42, n_init = 10)
clusters = kmeans.fit_predict(x_pca)

plt.figure(figsize = (6, 4))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = clusters)

plt.title('k-means clustering on pca data')
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')

plt.grid(alpha=0.3)
plt.show()





12] final interpretation
PCA was applied to reduce the dimensionality of the dataset while preserving most of the important information.
the first principal components capture the maximum variance present in the data.
the explained variance ratio shows how much information each principal component retains.
if the first two principal components explain a large portion of the total variance, then dimensionality reduction is effective.
the pca-transformed scatter plot helps visualize the data in reduced dimensions and may show better separation or structure compared to the original feature space.
overall, PCA simplifies the dataset, reduces complexity, and helps in visualization and further analysis.
```

# t-Distributed Stochastic Neighbor Embedding (t-SNE) Algorithm

```python
t-SNE is a non-linear dimensionality reduction technique used to visualize high-dimensional data in a lower-dimensional space, usually 2d or 3d, while preserving local similarity between data points.

how it works:
it measures how similar points are in the original high-dimensional space.
it tries to place similar points close together in a lower-dimensional space.
it uses probability distributions to preserve local neighborhoods.
it minimizes the difference between the original and reduced-space similarities.

    
    
    
    important parameters / methods:
n_components: number of output dimensions, usually 2 for visualization.
perplexity: controls the effective number of nearby neighbors considered.
random_state: ensures reproducible output.
standardization: a useful preprocessing step before applying t-sne.

    
    
    how to choose important parameters:
if the number of dimensions is given, use it directly.
otherwise, use 2 components for plotting.
choose perplexity usually between 5 and 50.
a common practical choice is perplexity = 30.
use random_state = 42 to get consistent results.

    
    
    
    real-world applications:
visualization of image datasets
handwritten digit exploration
gene expression data visualization
customer or document pattern exploration
exploratory analysis of high-dimensional datasets




advantages:
works well for non-linear data
preserves local structure effectively
gives clear 2d or 3d visualizations
useful for exploratory data analysis




limitations:
mainly useful for visualization, not direct prediction
output can change with different parameter values
does not preserve global distances well
can be slow for large datasets



1] import the packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE

import warnings
warnings.filterwarnings('ignore')



2] generate / obtain the dataset
# iris is a common and safe dataset for t-sne practical
iris = load_iris()

x = iris.data
y = iris.target

feature_names = iris.feature_names
target_names = iris.target_names

print("original data shape:", x.shape)





3] visualize raw data
plt.figure(figsize=(5, 4))

plt.scatter(x[:, 0], x[:, 1])

plt.title('raw data (before t-sne)')
plt.xlabel('feature 1')
plt.ylabel('feature 2')

plt.grid(alpha=0.3)
plt.show()





4] standardize the data
# standardization is a safe preprocessing step
# it prevents larger-scale features from dominating

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

df = pd.DataFrame(x_scaled, columns = feature_names)
df.head()





5] applying t-SNE
# t-sne reduces high-dimensional data into 2d for visualization
# perplexity is usually kept between 5 and 50
# 30 is a common exam-safe value

tsne = TSNE(n_components = 2, perplexity = 30, random_state = 42)
x_tsne = tsne.fit_transform(x_scaled)

print("t-sne transformed data shape:", x_tsne.shape)





6] create transformed dataframe
tsne_df = pd.DataFrame(x_tsne, columns = ['tsne1', 'tsne2'])
tsne_df['species'] = y

tsne_df.head()





7] visualize t-SNE output
# this shows how the data looks after t-sne embedding
# if similar classes form separate groups, visualization is meaningful

plt.figure(figsize=(6, 4))

for i in range(len(target_names)):
    plt.scatter( x_tsne[y == i, 0], x_tsne[y == i, 1], label=target_names[i] )

plt.title('data after t-sne')
plt.xlabel('t-sne component 1')
plt.ylabel('t-sne component 2')
plt.legend()

plt.grid(alpha=0.3)
plt.show()





8] compare original vs t-SNE
# this compares original feature space with t-sne space
# it helps us see whether class grouping becomes clearer after t-sne

plt.figure(figsize=(11, 5))

datasets = [
    (x, 'original data', 'feature 1', 'feature 2'),
    (x_tsne, 'data after t-sne', 't-sne component 1', 't-sne component 2')
]

for idx, (data, title, xlabel, ylabel) in enumerate(datasets):
    
    plt.subplot(1, 2, idx + 1)

    for i in range(len(target_names)):
        plt.scatter( data[y == i, 0], data[y == i, 1], label = target_names[i] )

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()






9] applying KNN on t-SNE output
from sklearn.metrics import silhouette_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# split data
x_train, x_test, y_train, y_test = train_test_split( x_tsne, y, test_size=0.3, random_state=42, stratify=y)

# k-nn model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

# metrics
sil_score = silhouette_score(x_tsne, y)

print("silhouette score:", round(sil_score, 4))
print("\nclassification report:")
print(classification_report(y_test, y_pred, target_names=target_names))

cm = confusion_matrix(y_test, y_pred)
print("confusion matrix:")
print(cm)





# plot knn predictions in t-sne space
plt.figure(figsize=(6, 4))

plt.scatter(x_test[:, 0], x_test[:, 1], c = y_pred)

plt.title('k-nn predictions on t-sne data')
plt.xlabel('t-sne component 1')
plt.ylabel('t-sne component 2')

plt.grid(alpha=0.3)
plt.show()





10] final interpretation
t-sne was applied to reduce the high-dimensional dataset into 2 dimensions for visualization.
the transformed plot helps reveal local structure by placing similar observations closer together.
if the classes or groups appear more clearly separated in the t-sne plot, it indicates that the embedding has captured meaningful patterns in the data.
t-sne is mainly used for visualization and exploratory analysis rather than direct prediction.
overall, t-sne helps represent complex high-dimensional data in a simpler 2d form that is easier to interpret.



```

# Independent Component Analysis (ICA) Algorithm

```python
ICA is a technique used to separate a multivariate signal into independent non-gaussian components. it is mainly used to recover original source signals from mixed observed signals.

how it works:
start with observed mixed signals.
assume that the observed data is a linear combination of hidden independent source signals.
estimate a transformation that makes the recovered components as statistically independent as possible.
the resulting components represent the separated source signals.

    
    
    important parameters / methods:
n_components: number of independent components to recover.
random_state: used for reproducible output.
fastica: a commonly used algorithm for performing ica efficiently.

    
    
    
    how to choose important parameters:
if the number of source signals is known, use that value as n_components.
if not given, choose a reasonable number based on the number of observed mixed signals.
use a fixed random_state such as 42 for consistent results.

    
    
    
    real-world applications:
audio source separation
speech enhancement
biomedical signal processing such as eeg and ecg
image separation
noise removal and signal recovery

    
    
    advantages:
can separate mixed signals into independent components
useful for blind source separation
works well when source signals are non-gaussian and independent
helpful in audio and biomedical applications

    
    
    
    limitations:
assumes source signals are statistically independent
may fail if signals are gaussian or highly dependent
recovered signals may differ in sign or scale
number of observed mixtures should usually be at least the number of source signals




PART 1: working on signals


1] import the packages
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import FastICA

import warnings
warnings.filterwarnings('ignore')



2] generate / obtain the dataset
# create two simple independent source signals
# these are easy to understand and safe for exam

np.random.seed(0)

time = np.linspace(0, 10, 1000)

s1 = np.sin(2 * time)                  # source signal 1: sine wave
s2 = np.sign(np.sin(3 * time))         # source signal 2: square wave

# combine source signals into one matrix
s = np.c_[s1, s2]

# standardize each source signal
s = s / s.std(axis = 0)

print("source signal shape:", s.shape)





3] mixing the signals
# create a simple mixing matrix
# this combines both source signals into observed mixed signals

a = np.array([ [1, 0.5], [0.5, 1] ])
x = s @ a.T

print("mixed signal shape:", x.shape)





4] applying ICA
# ica tries to recover the original independent components
# n_components = 2 because we have two source signals

ica = FastICA(n_components=2, random_state=42)
s_ica = ica.fit_transform(x)

print("recovered signal shape:", s_ica.shape)





5] comparing original, mixed and seperated signals
# this is the most important visual comparison for exam
# it helps us see whether ica successfully separated the signals

fig, axes = plt.subplots(3, 1, figsize=(10, 8))

axes[0].plot(time, s[:, 0], label = 'source 1')
axes[0].plot(time, s[:, 1], label = 'source 2')
axes[0].set_title('original source signals')
axes[0].legend()
axes[0].grid(alpha=0.3)

axes[1].plot(time, x[:, 0], label = 'mixed signal 1')
axes[1].plot(time, x[:, 1], label = 'mixed signal 2')
axes[1].set_title('mixed signals')
axes[1].legend()
axes[1].grid(alpha=0.3)

axes[2].plot(time, s_ica[:, 0], label = 'recovered signal 1')
axes[2].plot(time, s_ica[:, 1], label = 'recovered signal 2')
axes[2].set_title('separated signals (ica)')
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()





6] final interpretation
ICA was applied to separate the mixed signals into independent source components.
the mixed signals are linear combinations of the original source signals.
after applying ICA, the recovered signals represent the estimated independent components.
if the recovered signals are similar in pattern to the original signals, then the separation is successful.
small differences in sign or scale are normal in ica and do not affect the correctness of separation.
overall, ICA is useful for blind source separation where hidden independent signals need to be recovered from mixed observations.






PART 2: working on images

1] import libraries
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import FastICA
from skimage import data, color, transform



2] load and preprocess images
# load two sample images from skimage
img1 = color.rgb2gray(data.astronaut())
img2 = data.coins()

# resize images to same shape
img1 = transform.resize(img1, (256, 256))
img2 = transform.resize(img2, (256, 256))

print("image shape:", img1.shape)




3] visualize original images
plt.figure(figsize=(7, 4))

plt.subplot(1, 2, 1)
plt.imshow(img1, cmap = 'gray')
plt.title('original image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2, cmap = 'gray')
plt.title('original image 2')
plt.axis('off')

plt.show()





4] flatten and combine signals
# flatten images into vectors
s1 = img1.flatten()
s2 = img2.flatten()

# stack into source matrix
S = np.c_[s1, s2]

# standardize signals
S = S / S.std(axis=0)





5] mixing the signals & reshaping images
# create mixing matrix
A = np.array([ [1, 0.5], [0.5, 1] ])

# generate mixed signals
X = S @ A.T
mixed1 = X[:, 0].reshape(256, 256)
mixed2 = X[:, 1].reshape(256, 256)





6] visualize mixed images
plt.figure(figsize=(7, 4))

plt.subplot(1, 2, 1)
plt.imshow(mixed1, cmap='gray')
plt.title('mixed image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(mixed2, cmap='gray')
plt.title('mixed image 2')
plt.axis('off')

plt.show()






7] applying ICA
# apply fastica to recover original signals

ica = FastICA(n_components = 2, random_state = 42)
S_ica = ica.fit_transform(X)







8] reshape recovered images
rec1 = S_ica[:, 0].reshape(256, 256)
rec2 = S_ica[:, 1].reshape(256, 256)






9] visualize separated images
plt.figure(figsize=(7, 4))

plt.subplot(1, 2, 1)
plt.imshow(rec1, cmap='gray')
plt.title('separated image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rec2, cmap='gray')
plt.title('separated image 2')
plt.axis('off')

plt.show()







10] final interpretation
two original images were combined to form mixed images using a linear combination.
the mixed images contain information from both original images, making them difficult to distinguish individually.
independent component analysis (ICA) was applied to separate the mixed signals into independent components.
the recovered images resemble the original images, showing that ica successfully separated the sources.
differences in brightness, contrast, or sign are normal and do not affect the correctness of separation.
overall, ICA is effective for separating mixed image data into independent components, demonstrating its use in blind source separation.
```
