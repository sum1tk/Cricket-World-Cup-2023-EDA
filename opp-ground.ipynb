```python
#required library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('CWC23_all_innings.csv')


df_bat=df[df['bat_or_bowl']=='bat']
df_bowl=df[df['bat_or_bowl']=='bowl']
```

```python
teams=['AFG',
 'AUS',
 'BAN',              
'ENG',             
'IND',             
'NED',             
'NZ',             
'PAK',              
'SA',             
'SL'] 
```

```python


for i in teams:
    team_batting_performance=df_bat[df_bat['team']==i].groupby(['team', 'start_date', 'opposition']).agg({
        'runs': 'sum',
        '4s': 'sum',
        '6s': 'sum',
        'sr': 'mean',
        'runs_per_ball': 'mean'
    }).idxmax()
    print(i)
    with pd.option_context('display.max_rows',None):
        print(team_batting_performance)
    print('\n')



```

```python


for i in teams:
    team_bowling_performance=df_bowl[df_bowl['team']==i].groupby(['team', 'start_date', 'opposition']).agg({
    'wkts': 'sum',
    'econ': 'mean',
    'wicketball_prob': 'mean',
    'mdns': 'sum'
    }).idxmax()
    print(i)
    with pd.option_context('display.max_rows',None):
        print(team_bowling_performance)
    print('\n')
```

```python
#ground analysis
for i in teams:
    team_batting_performance=df_bat[df_bat['team']==i].groupby(['team', 'start_date', 'ground']).agg({
        'runs': 'sum',
        '4s': 'sum',
        '6s': 'sum',
        'sr': 'mean',
        'runs_per_ball': 'mean'
    }).idxmax()
    print(i)
    with pd.option_context('display.max_rows',None):
        print(team_batting_performance)
    print('\n')
```

```python

for i in teams:
    team_bowling_performance=df_bowl[df_bowl['team']==i].groupby(['team', 'start_date', 'ground']).agg({
    'wkts': 'sum',
    'econ': 'mean',
    'wicketball_prob': 'mean',
    'mdns': 'sum'
    }).idxmax()
    print(i)
    with pd.option_context('display.max_rows',None):
        print(team_bowling_performance)
    print('\n')
```

```python
#player 
players=df['player'].unique()
```

```python

for i in players:
    player_data = df_bowl[df_bowl['player'] == i]
    
    if not player_data.empty:
        team_bowling_performance = player_data.groupby(['player', 'start_date', 'opposition']).agg({
            'wkts': 'sum',
            'econ': 'mean',
            'wicketball_prob': 'mean',
            'mdns': 'sum'
        }).idxmax(skipna=True)
        
        print(i)
        with pd.option_context('display.max_rows', None):
            print(team_bowling_performance)
        print('\n')
    else:
        print(f"No data for player {i}\n")
```
