```python
#required library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('CWC23_all_innings.csv')
df['start_date'] = pd.to_datetime(df['start_date'])
display(df)

#dividing dataset 
df_bat=df[df['bat_or_bowl']=='bat']
df_bowl=df[df['bat_or_bowl']=='bowl']

display(df_bat)
display(df_bowl)
```

```python
df_bat=df_bat.drop(columns='player')
df_bowl=df_bowl.drop(columns='player')
```

```python
# Batting Average:
teams=sorted(df['team'].unique())

# The average number of runs scored per wicket lost, providing insight into batting consistency.
batting_avg=np.array(df_bat.groupby('team')['runs'].sum())/np.array(df_bowl.groupby('opposition')['wkts'].sum())
team_batting_avg={'team':teams,'batting_avg':batting_avg}
df_batting_avg=pd.DataFrame(team_batting_avg).sort_values(by='batting_avg')
display(df_batting_avg)

sns.set_style('darkgrid')
fig,ax=plt.subplots(1,figsize=(17,7))

ax.barh(df_batting_avg['team'],df_batting_avg['batting_avg'])
ax.set_ylabel('Team')
ax.set_xlabel('Batting Average')
```

```python
# Percentage of 4s and 6s:
# Reflects the team's ability to score boundaries and maintain a healthy run rate
perc_4=((df_bat.groupby('team')['4s'].sum()/df_bat['4s'].sum())*100).sort_values()

perc_6=((df_bat.groupby('team')['6s'].sum()/df_bat['6s'].sum())*100).sort_values()

display(perc_4)
display(perc_6)


fig,ax=plt.subplots(1,2,figsize=(17,7))


ax[0].bar(perc_4.keys(),perc_4)
ax[0].set_xlabel('Team')
ax[0].set_ylabel('Percentage of 4s')

ax[1].bar(perc_6.keys(),perc_6)
ax[1].set_xlabel('Team')
ax[1].set_ylabel('Percentage of 6s')
```

```python
# Percentage of Centuries/Fifties:

# Indicates how frequently the team's batsmen convert starts into significant scores.
centuries=df_bat.groupby('team')['runs'].apply(lambda x: x[x>=100].count()).sort_values()
half_centuries=df_bat.groupby('team')['runs'].apply(lambda x: x[(x >= 50) & (x < 100)].count()).sort_values()

fig,ax=plt.subplots(1,2,figsize=(17,7))

display(centuries)
display(half_centuries)

ax[0].bar(centuries.keys(),centuries)
ax[0].set_xlabel('Team')
ax[0].set_ylabel('centuries')

ax[1].bar(half_centuries.keys(),half_centuries)
ax[1].set_xlabel('Team')
ax[1].set_ylabel('half centuries')
```

```python
# Bowling Average:

# The average number of runs conceded per wicket taken, indicating the effectiveness of the bowling unit.
df_bowling_avg=df_bowl.groupby('team')['runs'].sum()/df_bowl.groupby('team')['wkts'].sum()
df_bowling_avg=df_bowling_avg.sort_values(ascending=False)
display(df_bowling_avg)
fig,ax=plt.subplots(1,figsize=(17,7))
ax.barh(df_bowling_avg.keys(),df_bowling_avg)
ax.set_ylabel('Team')
ax.set_xlabel('Bowling Average')


plt.show()
```

```python
econ_avg=df_bowl.groupby('team')['econ'].mean().sort_values(ascending=False)
mdns_total=df_bowl.groupby('team')['mdns'].sum().sort_values()

display(econ_avg)
display(mdns_total)


fig,ax=plt.subplots(1,2,figsize=(17,7))


ax[0].bar(econ_avg.keys(),econ_avg)
ax[0].set_xlabel('Team')
ax[0].set_ylabel('Economy average')

ax[1].bar(mdns_total.keys(),mdns_total)
ax[1].set_xlabel('Team')
ax[1].set_ylabel('Total Maidens')
```
